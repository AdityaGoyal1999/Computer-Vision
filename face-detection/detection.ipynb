## Face Detector - Viola Jones

Task: Detect all of the faces in the image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./images/faces.jpeg')
face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') 
```

```python
face_boxes = face_detector.detectMultiScale(img)
```

```python
for (x,y,w,h) in face_boxes: # for loop to detect multiple faces
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0), 5) 

plt.figure(figsize=(7,7))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.xticks([]), plt.yticks([]);
```

Note: As we can see Viola Jones face detector is not quite robust in detecting images. There are much better ways of doing so.
